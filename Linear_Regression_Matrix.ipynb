```python
import numpy as np

# Define the design matrix X and response vector Y (from the table)
X = np.array([
    [1, 0.5, 0.4],
    [1, 0.8, 0.6],
    [1, 0.9, 0.7],
    [1, 1.1, 1.2],
    [1, 1.3, 1.4],
    [1, 1.4, 1.7]
])

Y = np.array([0.19, 0.28, 0.30, 0.25, 0.29, 0.28]).reshape(-1, 1)  # Ensure column vector

# Step 1: Compute X^T * X
XT = X.T
A = XT @ X

# Step 2: Compute inverse of X^T * X (K)
try:
    K = np.linalg.inv(A)
except np.linalg.LinAlgError:
    print("Matrix is singular; cannot compute inverse.")
    exit()

# Step 3: Calculate least squares coefficients B = (X^T X)^-1 X^T Y
B = K @ XT @ Y

# Step 4: Compute predicted values (M) and errors (E)
M = X @ B
E = Y - M

# Step 5: Find maximum absolute error
MaxErr = np.max(np.abs(E))

# Print results
print("Matrix X^T X:")
print(A, "\n")

print("Inverse of X^T X (K):")
print(K, "\n")

print("Coefficients B (β):")
print(B.flatten(), "\n")  # Flatten to show as row vector

print("Predicted values M:")
print(M.flatten(), "\n")

print("Errors E:")
print(E.flatten(), "\n")

print(f"Maximum absolute error: {MaxErr:.4f}")
```

```python
# prompt: scatter plot for predicted values

import matplotlib.pyplot as plt

# Create a scatter plot of the actual values (Y) versus the predicted values (M)
plt.figure(figsize=(8, 6))
plt.scatter(Y, M, alpha=0.7)
plt.plot(Y, Y, color='red', linestyle='--', label='Ideal Fit (Y=M)') # Add a line for the ideal fit
plt.xlabel('Actual Values (Y)')
plt.ylabel('Predicted Values (M)')
plt.title('Scatter Plot of Actual vs. Predicted Values')
plt.grid(True)
plt.legend()
plt.show()
```
